```python
import cv2


#####DIKDORTGEN CİZMEK İCİN#####
def dikdortgen_ciz(event,x,y,flag,param):
    global pt1,pt2,topLeft_clicked,bottomRight_clicked

    if event==cv2.EVENT_LBUTTONDOWN: #sol tusla basildiginda

        if topLeft_clicked==True and bottomRight_clicked==True: #eger iki nokta da secildiyse varsayilan degerler atanir
            pt1=(0,0) #noktalari sifirlar
            pt2 =(0,0)
            topLeft_clicked = False 
            bottomRight_clicked = False

        if topLeft_clicked==False: #eger ilk nokta secilmediyse
            pt1=(x,y) #ilk noktayi alir
            topLeft_clicked=True #ilk noktayi secildi olarak isaretler

        elif bottomRight_clicked==False: #eger ilk nokta secildiyse ve ikinci nokta secilmediyse
            pt2=(x,y)
            bottomRight_clicked=True



pt1=(0,0) #noktalari sifirlar
pt2 =(0,0)

topLeft_clicked = False  #noktalarin secilip secilmedigini kontrol eder
bottomRight_clicked = False



cap = cv2.VideoCapture(0) #kamerayi acar

cv2.namedWindow("Test")
cv2.setMouseCallback("Test",dikdortgen_ciz)


while True:
    ret,frame = cap.read() #kameradan goruntuyu alir

    ##Sekilleri Cizdirme##

    if topLeft_clicked==True: #eger ilk nokta secildiyse

        cv2.circle(frame,center=pt1,radius=5,color=(0,0,255),thickness=-1)

    if topLeft_clicked and bottomRight_clicked: #eger iki nokta secildiyse

        cv2.rectangle(frame,pt1,pt2,(0,0,255),3)


    cv2.imshow('Test',frame) #goruntuyu ekrana yansitir

    if cv2.waitKey(1) & 0xFF==ord('q'): #q tusuna basilinca donguden cikar
        break

cap.release() #kamerayla baglantiyi koparir
cv2.destroyAllWindows() #pencereleri kapatir

```
