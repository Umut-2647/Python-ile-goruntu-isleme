```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('media/internal_external.jpg',0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
img.shape
```

```python
contours, hierarchy = cv2.findContours(img,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
type(contours)
```

```python
len(contours)
```

```python
type(hierarchy)
```

```python
hierarchy.shape
```

```python
hierarchy
```

```python
external_contours =np.zeros(img.shape)
```

```python
plt.imshow(external_contours)
```

```python
for i in range(len(contours)):
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(external_contours,contours,i,255,-1)
```

```python
plt.imshow(external_contours,cmap='gray') #noktalarn dndaki alanlar beyaz yapar
```

```python
image_internal =np.zeros(img.shape)

for i in range(len(contours)):
    if hierarchy[0][i][3] != -1:
        cv2.drawContours(image_internal,contours,i,255,-1)
```

```python
plt.imshow(image_internal,cmap='gray') #noktalarn iindeki alanlar beyaz yapar
```
