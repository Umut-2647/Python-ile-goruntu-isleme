```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('media/face_dog.png')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
```

```python
plt.imshow(edges)
```

```python
blur = cv2.blur(img,(5,5)) #fotograffaki parazitlemeyi azaltmak icin kullanilir
```

```python
edges =cv2.Canny(image=blur, threshold1=100, threshold2=180)
```

```python
plt.imshow(edges)
```
