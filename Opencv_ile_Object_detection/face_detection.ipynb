```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
adam=cv2.imread('media/adam.jpeg',0)
plt.imshow(adam,cmap='gray')
```

```python
kadin=cv2.imread('media/kadin.jpg',0)
plt.imshow(kadin,cmap='gray')
```

```python
toplu_foto=cv2.imread('media/toplu_foto.jpg',0)
plt.imshow(toplu_foto,cmap='gray')
```

```python
face_cascade =cv2.CascadeClassifier("media/frontalface.xml") #yuz tanimlayacagimiz xml dosyasi
```

```python
def detect_face(img):
    face_img = img.copy()

    face_res =face_cascade.detectMultiScale(face_img) #yuz tespiti yapar

    for (x,y,w,h) in face_res: #yuzun koordinatlarini alir
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,0,0),4) #yuzun etrafina dikdortgen cizer

    return face_img
```

```python
result = detect_face(adam)
```

```python
plt.imshow(result,cmap='gray')
```

```python
result = detect_face(kadin)
plt.imshow(result,cmap='gray')
```

```python
result= detect_face(toplu_foto)
plt.imshow(result,cmap='gray')
```

```python
eye_cascade=cv2.CascadeClassifier("media/eye.xml") #goz tanimlayacagimiz xml dosyasi
```

```python
def detect_eye(img):
    face_img = img.copy()

    eyes_res =eye_cascade.detectMultiScale(face_img) #yuz tespiti yapar

    for (x,y,w,h) in eyes_res: #yuzun koordinatlarini alir
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,0,0),4) #yuzun etrafina dikdortgen cizer

    return face_img
```

```python
result = detect_eye(adam)
plt.imshow(result,cmap='gray')
```

```python
result = detect_eye(kadin)
plt.imshow(result,cmap='gray')
```

```python
result= detect_eye(toplu_foto)
plt.imshow(result,cmap='gray')
```
