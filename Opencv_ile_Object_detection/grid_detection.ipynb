```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('media/flat_chessboard.jpg')
```

```python
plt.imshow(flat_chess)
```

```python
#birinci deger konturu bulup bulamadigini soyler
found ,corners = cv2.findChessboardCorners(flat_chess,(7,7))
```

```python
found
```

```python
if found:
    print("Opencv was able to find the corners")
else:
    print("OpenCV did not find corners. Double check your patternSize.")
    
```

```python
corners
```

```python
corners.shape
```

```python
flat_chess_copy = flat_chess.copy()
cv2.drawChessboardCorners(flat_chess_copy,(7,7),corners,found)
```

```python
plt.imshow(flat_chess_copy)
```

```python
dots = cv2.imread('media/dot_grid.jpg')
plt.imshow(dots)
```
