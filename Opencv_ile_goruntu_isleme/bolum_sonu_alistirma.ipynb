```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img,cmap=None): #resimleri gosteren fonksiyon
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
path = os.path.join(os.getcwd(), "media", "zurafa.jpg") #resmin yolunu belirliyoruz
img=cv2.imread(path) #resmi okuyoruz
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #resmi RGB formatına çeviriyoruz
```

```python
display_img(img)
```

```python
img=cv2.imread(path,0) #0 siyah beyaz yapar
ret,thresh1=cv2.threshold(img,127,255,cv2.THRESH_BINARY) #127 den kucuk olanlari 0 yapar, 127 den buyuk olanlari 255 yapar
display_img(thresh1,cmap='gray')  #siyah beyaz yapar    
```

```python
img=cv2.imread(path)
hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV) #hsv donusumu
display_img(hsv)
```

```python
kernel = np.ones((4,4),dtype=np.float32) /10 #4x4 0.1 degerine sahip kernel olusturduk
```

```python
kernel
```

```python
img=cv2.imread(path) #resmi okuyoruz
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #resmi RGB formatına çeviriyoruz
```

```python
dst = cv2.filter2D(img,-1,kernel) #resmi filtreliyoruz
```

```python
display_img(dst)
```

```python
img=cv2.imread(path,0) #0 siyah beyaz yapar

sobel_x=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5) #yatay sobel filtresi
#morfolojik islemler
```

```python
display_img(sobel_x,cmap='gray')
```

```python
img=cv2.imread(path)
color = ('b','g','r')

for i,col in enumerate(color): #histogram cizimi
    histr=cv2.calcHist([img],[i],None,[256],[0,256]) #histogram hesaplama
    plt.plot(histr,color=col) #histogram cizimi
    plt.xlim([0,256]) #x eksenini belirleme

plt.title("HISTOGRAM")
plt.show()
```
