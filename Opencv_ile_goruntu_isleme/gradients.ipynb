```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img): #resimleri gosteren fonksiyon
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap="gray")
```

```python
path = os.path.join(os.getcwd(), "media", "sudoku2.jpg") #resimlerin yolunu belirleme
img = cv2.imread(path,0) #gri tonlarda okuma
```

```python
display_img(img)
```

```python
# x ekseninde yapmak istedignimizde 1,0 y ekseninde yapmak istedigimizde 0,1
sobel_x=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5) #dikey cizgilerin daha belirgin olamasini sagliyoruz
```

```python
display_img(sobel_x)
```

```python
sobel_y=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5) #yatay cizgilerin daha belirgin olamasini sagliyoruz
display_img(sobel_y)
```

```python
laplacian = cv2.Laplacian(img,cv2.CV_64F) #hem dikey hem yatay cizgilerin belirgin olmasini sagliyoruz
display_img(laplacian)
```

```python
blended =cv2.addWeighted(sobel_x,0.5,sobel_y,0.5,0) #sobel_x ve sobel_y yi birlestiriyoruz
```

```python
display_img(blended)
```

```python
ret , th1= cv2.threshold(img,50,255, cv2.THRESH_BINARY) #resmi threshold etme
display_img(th1)
```

```python
kernel = np.ones((4,4),np.uint8)
```

```python
gradient =cv2.morphologyEx(blended,cv2.MORPH_GRADIENT,kernel) #gradient
display_img(gradient)
```
