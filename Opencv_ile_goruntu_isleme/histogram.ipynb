```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path_horse=os.path.join('media','at.jpg') #resimlerin yolunu aliyoruz
path_duvar=os.path.join('media','duvar.jpg')
path_gokkusagi=os.path.join('media','gokkusagi.jpg')

dark_horse=cv2.imread(path_horse) #resimleri okuyoruz
duvar=cv2.imread(path_duvar) #resimleri okuyoruz
gokkusagi=cv2.imread(path_gokkusagi) #resimleri okuyoruz




show_horse=cv2.cvtColor(dark_horse,cv2.COLOR_BGR2RGB) #resimleri RGB ye ceviriyoruz
show_duvar=cv2.cvtColor(duvar,cv2.COLOR_BGR2RGB) #resimleri RGB ye ceviriyoruz
show_gokkusagi=cv2.cvtColor(gokkusagi,cv2.COLOR_BGR2RGB) #resimleri RGB ye ceviriyoruz
```

```python
img = duvar
```

```python
color = ('b','g','r')

for i,col in enumerate(color):
    histr=cv2.calcHist(images=[img],channels=[i],mask=None,histSize=[256],ranges=[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,250])

plt.title("HISTOGRAM")
```
