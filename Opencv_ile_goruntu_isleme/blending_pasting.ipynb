```python
img2gray=cv2.cvtColor(img1_rgb, cv2.COLOR_RGB2GRAY) #resmi griye cevirme
plt.imshow(img2gray, cmap="gray") #resmi gosterme
```

```python
mask_inv=cv2.bitwise_not(img2gray) #resmi ters cevirme
plt.imshow(mask_inv,cmap="gray") #resmi gosterme
```

```python
mask_inv.shape #resmin boyutunu gosterme
```

```python
white_background=np.full((800,800,3),255,dtype=np.uint8) #beyaz bir arka plan olusturma
white_background.shape #resmin boyutunu gosterme
```

```python
back=cv2.bitwise_or(white_background,white_background,mask=mask_inv) #resimleri birlestirme
```

```python
back.shape #resmin boyutunu gosterme
```

```python
plt.imshow(back) #resmi gosterme
```

```python
fg=cv2.bitwise_or(img1_rgb,img1_rgb,mask=mask_inv) #resimleri birlestirme
```

```python
plt.imshow(fg) #resmi gosterme
```

```python
roi= img_rgb[0:800, 200:1000] #resimden bir bolum secme 
```

```python
final_roi=cv2.bitwise_or(roi,fg) #resimleri birlestirme
plt.imshow(final_roi) #resmi gosterme
```

```python
large_img=img_rgb #buyuk resim
small_img=final_roi #kucuk resim
```

```python
large_img[0:800, 200:1000]=small_img #kucuk resmi buyuk resmin uzerine yerlestir
plt.imshow(large_img) #resmi gosterme
```
