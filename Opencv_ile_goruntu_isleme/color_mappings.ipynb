```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import os
import cv2
```

```python
path = os.path.join(os.getcwd(), "media", "manzara.jpeg")
img = cv2.imread(path)
```

```python
plt.imshow(img)  #BGR formatında gösterir
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)  #RGB formatında gösterir
```

```python
img=cv2.cvtColor(img,cv2.COLOR_RGB2HSV) #daha cok mavi agirlikli bir resim olusturur
plt.imshow(img)  #HSV formatında gösterir
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2HLS) #daha cok mor agirlikli bir resim olusturur
plt.imshow(img)  #HLS formatında gösterir
```
