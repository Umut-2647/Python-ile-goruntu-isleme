```python
path = os.path.join(os.getcwd(), "media", "bulmaca.jpg") #resimlerin yolunu belirleme
img = cv2.imread(path,0) #resmi okuma
plt.imshow(img,cmap="gray") #resmi ekrana bastırma
```

```python
def show_pic(img):
    fig=plt.figure(figsize=(15,15)) #kaca kaclik bir olcekleme yapilacagini belirleme
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap="gray")
```

```python
show_pic(img)
```

```python
ret,th=cv2.threshold(img,127,255,cv2.THRESH_BINARY) #resmi threshold etme
show_pic(th)
```

```python
path = os.path.join(os.getcwd(), "media", "sudoku.jpeg") #resimlerin yolunu belirleme
img_1=cv2.imread(path,0) #resmi okuma
plt.imshow(img_1,cmap="gray") #resmi ekrana bastırma
```

```python
show_pic(img_1)
```

```python
ret,th1=cv2.threshold(img_1,127,255,cv2.THRESH_BINARY) #resmi threshold etme
show_pic(th1)
```

```python
th2=cv2.adaptiveThreshold(img_1,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8) #resmin ortalama thresh degerini belirler
show_pic(th2)
```
