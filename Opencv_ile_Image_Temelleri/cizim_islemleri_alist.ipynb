```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
```

```python

path = os.path.join(os.getcwd(), "media", "manzara.jpeg")
img =cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
#kare cizilmesi
```

```python
cv2.rectangle(img,pt1=(75,70),pt2=(120,90),color=(255,0,0),thickness=2)
plt.imshow(img)
```

```python
cv2.flip(img,0)
plt.imshow(img)
```

```python
plt.imshow(img)
```

```python
noktalar=np.array([[200,20],[225,60],[175,60]],dtype=np.int32)
cv2.polylines(img,[noktalar],isClosed=True,color=(0,0,255),thickness=2)
plt.imshow(img)
```

```python
def daire_ciz(event,x,y,flags,param):
    if event==cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),50,(0,0,255),2)
    elif event==cv2.EVENT_RBUTTONDBLCLK:
        cv2.circle(img,(x,y),50,(0,255,0),2)

cv2.namedWindow("Image")

img=cv2.imread(path)
cv2.setMouseCallback("Image",daire_ciz)

while True:
    cv2.imshow("Image",img)
    if cv2.waitKey(1) & 0xFF==27:
        break

cv2.destroyAllWindows()
```
