```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bos_foto=np.zeros(shape=(512,512,3),dtype=np.uint16) #512 512 boyutlarinda siyah bir resim olusturduk
plt.imshow(bos_foto)
```

```python
cv2.putText(bos_foto,text="Merhaba",org=(10,500),
            fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=2,color=(255,255,255),thickness=2)
plt.imshow(bos_foto) 
```

```python
noktalar=np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
noktalar.shape
```

```python
cv2.polylines(bos_foto,[noktalar],isClosed=True,color=(255,0,0),thickness=5)
#isClosed=True ile noktalar arasina cizgi cekildi
plt.imshow(bos_foto)
```
